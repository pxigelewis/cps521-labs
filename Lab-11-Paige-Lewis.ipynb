```python
from networkx.algorithms.community.centrality import girvan_newman
from networkx.algorithms.community import greedy_modularity_communities
import networkx as nx
import community as community_louvain
import matplotlib.pyplot as plt
```

```python
nodes = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I']
edges = [('A','B'), ('A','C'), ('B','C'), ('B','H'), ('C','D'), ('H','I'), ('H','G'), ('D','E'), ('D','F'), ('I','G'), ('G','E'), ('E','F')]

graph = nx.Graph()

graph.add_nodes_from(nodes)
graph.add_edges_from(edges)
nx.draw(graph, with_labels=True, font_weight='bold', node_color='yellow', node_size=700)
```

```python
communities = list(girvan_newman(graph))

node_groups = []
for com in communities[1]:
    node_groups.append(list(com))

print(node_groups)

color_map = []
for node in graph:
    if node in node_groups[0]:
        color_map.append('blue')
    elif node in node_groups[1]:
        color_map.append('red')
    else: 
        color_map.append('yellow')  
nx.draw(graph, node_color=color_map, with_labels=True)
plt.show()
```

```python
from collections import defaultdict
partitions = community_louvain.best_partition(graph)

node_groups = defaultdict(list)
for key, val in partitions.items():
    node_groups[val].append(key)

node_groups = list(node_groups.values())
print(node_groups)

color_map = []
for node in graph:
    if node in node_groups[0]:
        color_map.append('blue')
    elif node in node_groups[1]:
        color_map.append('red')
    else: 
        color_map.append('yellow')  
nx.draw(graph, node_color=color_map, with_labels=True)
plt.show()
```
