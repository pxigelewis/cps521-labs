```python
import pandas as pd

import numpy as np

from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score

import pydotplus
from IPython.display import Image
```

```python
df = pd.read_csv('heart-disease-dataset2.csv')

for col in df.columns:
    df = df[df[col] != '?']

x_cols = [col for col in df.columns if col != 'result']
X = df[x_cols]

Y = df['result'].values
```

```python
dt_gini = DecisionTreeClassifier(criterion='gini', random_state=2)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)

dt_train = dt_gini.fit(X_train, Y_train)

Y_pred = dt_train.predict(X_test)

dot_data = export_graphviz(dt_train, out_file=None, feature_names=x_cols, class_names=['0','1','2','3','4'], rounded=True, filled=True)

graph = pydotplus.graph_from_dot_data(dot_data)

accuracy = accuracy_score(Y_test, Y_pred)

print(f'Accuracy Score: {round(accuracy * 100,2)} %')

Image(graph.create_png())
```

```python
# Professor said to use Entropy Criterion with max_depth=3
dt_gini = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=2)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)

dt_train = dt_gini.fit(X_train, Y_train)

Y_pred = dt_train.predict(X_test)

dot_data = export_graphviz(dt_train, out_file=None, feature_names=x_cols, class_names=['0','1','2','3','4'], rounded=True, filled=True)

graph = pydotplus.graph_from_dot_data(dot_data)

accuracy = accuracy_score(Y_test, Y_pred)

print(f'Accuracy Score: {round(accuracy * 100,2)} %')

Image(graph.create_png())
```

```python
dt_gini = DecisionTreeClassifier(criterion='entropy', random_state=2)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)

dt_train = dt_gini.fit(X_train, Y_train)

Y_pred = dt_train.predict(X_test)

dot_data = export_graphviz(dt_train, out_file=None, feature_names=x_cols, class_names=['0','1','2','3','4'], rounded=True, filled=True)

graph = pydotplus.graph_from_dot_data(dot_data)

accuracy = accuracy_score(Y_test, Y_pred)

print(f'Accuracy Score: {round(accuracy * 100,2)} %')

Image(graph.create_png())
```
